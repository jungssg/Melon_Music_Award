```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv("data/melon_weekly_24_25_total.csv")

# 2. '연도'와 '주차'의 시작일을 합쳐서 실제 날짜 컬럼 생성
df['시작일_문자열'] = df['연도'].astype(str) + "." + df['주차'].str.split('~').str[0].str.strip()
df['마감일_문자열'] = df['연도'].astype(str) + "." + df['주차'].str.split('~').str[1].str.strip()
# 3. 날짜 형식으로 변환 (errors='coerce'는 잘못된 형식이 있을 경우 NaT로 처리)
df['시작일'] = pd.to_datetime(df['시작일_문자열'], format='%Y.%m.%d', errors='coerce')
df['마감일'] = pd.to_datetime(df['마감일_문자열'], format='%Y.%m.%d', errors='coerce')
# 4. 심사 기간 설정
# 2024.10.31~2025.11.19 기간이 하루라도 포함된 주간차트 사용
start_date = pd.to_datetime('2024-10-28') 
end_date = pd.to_datetime('2025-11-19')

# 5. 기간 내 데이터 필터링
df_mma_period = df[(df['시작일'] >= start_date) & (df['시작일'] <= end_date)]

# 6. 임시로 만든 컬럼 삭제 및 저장
df_final = df_mma_period.drop(columns=['시작일_문자열', '시작일'])
df_final.to_csv("mma_2025_period.csv", index=False, encoding="utf-8-sig")

print(f"추출된 주차 범위: {df_final['연도'].min()}년 {df_final[df_final['연도']==2024]['주차'].min()} ~ {df_final['연도'].max()}년 {df_final[df_final['연도']==2025]['주차'].max()}")
print(f"총 데이터 수: {len(df_final)}건")
```
