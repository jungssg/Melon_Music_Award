```python

def parse_activity_type(x):
    if pd.isna(x):
        return pd.Series(['unknown', 'unknown'])
    
    parts = [p.strip() for p in x.split(',')]
    
    gender = 'unknown'
    type_ = 'unknown'
    
    for p in parts:
        # 성별
        if p == '여성':
            gender = '여성'
        elif p == '남성':
            gender = '남성'
        elif p == '혼성':
            gender = '혼성'
        
        # 구성
        elif p == '솔로':
            type_ = '솔로'
        elif p == '그룹':
            type_ = '그룹'
    
    return pd.Series([gender, type_])

# 적용
artist_crawl[['성별', '구성']] = artist_crawl['활동유형'].apply(parse_activity_type)
```

### 앨범 발매일 추가수집

```python
album_data1 = pd.read_csv('album_release_dates.csv')
album_data2 = pd.read_csv('album_release_dates2.csv')
album_data3 = pd.read_csv('album_release_failures2.csv')
album_data4 = pd.read_csv('unmatched_re.csv')
```

```python
# album_data1.info()
# album_data2.info()
# album_data3.info()
album_data4.info()
```

```python
album_data = pd.concat([album_data1,album_data2, album_data3, album_data4],ignore_index=True)
album_data
```

```python
album_data['artist'] = album_data['artist'].replace(['오전 2:00', '2am', '2AM'], '2AM')
album_data['artist'] = album_data['artist'].replace(['오후 2:00', '2pm', '2PM'], '2PM')
```

```python
album_data['album'] = album_data['album'].replace("키스 먼저 할까요?' OST Part.3", "'키스 먼저 할까요?' OST Part.3")
```

```python
album_data[album_data['artist'] == '폴킴']
```

```python
album_data.to_csv('album_data.csv', index=False, encoding='utf-8-sig')
```
