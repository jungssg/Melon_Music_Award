```python

def parse_activity_type(x):
    if pd.isna(x):
        return pd.Series(['unknown', 'unknown'])
    
    parts = [p.strip() for p in x.split(',')]
    
    gender = 'unknown'
    type_ = 'unknown'
    
    for p in parts:
        # 성별
        if p == '여성':
            gender = '여성'
        elif p == '남성':
            gender = '남성'
        elif p == '혼성':
            gender = '혼성'
        
        # 구성
        elif p == '솔로':
            type_ = '솔로'
        elif p == '그룹':
            type_ = '그룹'
    
    return pd.Series([gender, type_])

# 적용
artist_crawl[['성별', '구성']] = artist_crawl['활동유형'].apply(parse_activity_type)
```

### df_crawl + artist_info

```python
# 컬럼명 통일 (df_crawl의 'artist'를 '아티스트명'으로 변경)
df_crawl = df_crawl.rename(columns={'artist': '아티스트명', '데뷔(곡)': '데뷔'})

# 데이터 합치기 (위아래로 연결)
combined_df = pd.concat([artist_info, df_crawl], ignore_index=True)
```

```python
combined_df = combined_df.drop(columns=['활동유형', '활동년대'])
combined_df.head()
```
