```python
# melon_2021_1.csv + 빠진 9월,10월 데이터 (melon_2021_ex.csv) 합치기
import pandas as pd

# 1. 두 파일 불러오기
df_main = pd.read_csv('/Users/js/MMA/Melon_Music_Award/data/melon_2021_1.csv')
df_ex = pd.read_csv('/Users/js/MMA/Melon_Music_Award/data/melon_2021_ex.csv')

# 2. 두 데이터프레임 연결
df_combined = pd.concat([df_main, df_ex], ignore_index=True)

# 3. 중복 데이터 제거
df_combined = df_combined.drop_duplicates()

# 4. 주차별(week_start) 및 순위(rank) 순서로 정렬
df_combined['week_start'] = pd.to_datetime(df_combined['week_start'])
df_combined = df_combined.sort_values(by=['week_start', 'rank'])

# 5. 저장
df_combined.to_csv('/Users/js/MMA/Melon_Music_Award/data/melon_2021.csv', index=False, encoding='utf-8-sig')

print("데이터 병합 및 정렬 완료!")
```
